# 01 — Build KITTI-C Manifest

Scan the KITTI-C dataset directory and build a CSV manifest used by all later notebooks.

**Requires:** KITTI-C data downloaded to the path in `configs/dataset_paths.yaml`

```python
import sys
from pathlib import Path

PROJECT_ROOT = None
for candidate in [Path.cwd().resolve(), *Path.cwd().resolve().parents]:
    if (candidate / 'configs' / 'dataset_paths.yaml').exists():
        PROJECT_ROOT = candidate
        break

if PROJECT_ROOT is None:
    raise RuntimeError('Could not locate project root from the current working directory')

if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))
```

```python
from src.utils.paths import get_dataset_path, get_output_path

kitti_c_root = get_dataset_path('kitti_c', 'root')
kitti_gt_root = get_dataset_path('kitti_c', 'gt_path')
manifests_dir = get_output_path('manifests')

print(f'KITTI-C root: {kitti_c_root}')
print(f'KITTI GT:     {kitti_gt_root}')
print(f'Manifests:    {manifests_dir}')
print(f'KITTI-C exists: {kitti_c_root.exists()}')
```

```python
from src.datasets.kitti_c import build_manifest

df = build_manifest(kitti_c_root, kitti_gt_root, split='test')
print(f'Manifest shape: {df.shape}')
df.head()
```

```python
print('Corruption types:', sorted(df['corruption_type'].unique()))
print('Severity levels: ', sorted(df['severity'].unique()))
print('Sequences:       ', sorted(df['sequence'].unique())[:5], '...')
print('GT available:    ', df['gt_path'].notna().sum(), '/', len(df))
```

```python
import pandas as pd

out_path = manifests_dir / 'kitti_c_manifest.csv'
df.to_csv(out_path, index=False)
print(f'Manifest saved to {out_path}')
```
